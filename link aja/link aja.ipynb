```python
from google_play_scraper import Sort, reviews_all

result = reviews_all(
    'com.telkom.mwallet',
    sleep_milliseconds=2000,
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=10000,  # Tentukan jumlah maksimum yang diinginkan
    filter_score_with=None
)

# Batasi hasil yang diambil menjadi 1000 ulasan
result = result[:10000]
```

```python
df_link = pd.DataFrame(np.array(result),columns=['review'])
df_link = df_link.join(pd.DataFrame(df_link.pop('review').tolist()))
df_link.head()
```
